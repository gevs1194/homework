```python
numbers=input("input all numbers separated by commas")
def average(numbers):
    return  float(sum(numbers))/len(numbers)
print average(numbers)
```

```python
numbers=input("input all numbers separated by commas")
argument=input("input argument")
def average(numbers):
    return  float(sum(numbers[-(argument):]))/float(int(argument))
print average(numbers)
```
